```python

from IPython.display import clear_output
boat_side= 'right'
MR=3
CR=3
ML=0
CL=0

print('\U0001f482'*ML, '\U0001f479'*CL, '\U0001f30a'*5, '\U0001f6A2', '\U0001f482'*MR, '\U0001f479'*CR)

while True:
    M=int(input())
    C=int(input())
    if M+C>2 or M+C<1:
      print('Invalid Move')
      continue

    else:
        print('valid move')

    if boat_side=='right':
      print('right')
      if M>MR or C>CR:
        print('invalid Move')
        continue
      else:
        print('valid')
        MR=MR-M
        CR=CR-C
        ML=ML+M
        CL=CL+C
        boat_side='left'
        print('\U0001f482'*ML, '\U0001f479'*CL,'\U0001f6A2', '\U0001f30a'*5, '\U0001f482'*MR, '\U0001f479'*CR)


    elif boat_side=='left':
      print('left')
      if M>ML or C>CL:
        print('invalid Move')
        continue
      else:
        print('valid')
        ML=ML-M
        CL=CL-C
        MR=MR+M
        CR=CR+C
        boat_side='right'

        print('\U0001f482'*ML, '\U0001f479'*CL, '\U0001f30a'*5,'\U0001f6A2', '\U0001f482'*MR, '\U0001f479'*CR)

    if MR<CR and MR>0:
      print('You Lose')
      break

    elif ML<CL and ML>0:
      print('You Lose')
      break
    if ML==3 and CL==3:
      print('You Win')
      break
```
